```r
library(dplyr)
library(survival)
library(survminer)
```

```r
pth <- '../data/TARGET-osteo-THC0-survival.tsv'

surv <- read.table(pth,
                  sep='\t',
                  header=T)

surv <- dplyr::as_tibble(surv)
```

```r
fit <- survfit(Surv(OS, vital) ~ cluster, data=surv)
g <- ggsurvplot(fit, conf.int=F, pval=T, risk.table=TRUE)
ggsave( "../img/TARGET-osteo-THC0-OS-survival.png", 
       print(g), 
       dpi = 1000 )
g
```

```r
fit <- survfit(Surv(EFS, event) ~ cluster, data=surv)
g <- ggsurvplot(fit, conf.int=F, pval=T, risk.table=TRUE)
ggsave( "../img/TARGET-osteo-THC0-EFS-survival.png", 
       print(g), 
       dpi = 1000 )
g
```
