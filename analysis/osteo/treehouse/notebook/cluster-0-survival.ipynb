```r
library(dplyr)
library(survival)
library(survminer)
```

```r
pth <- '../data/Tree-TARGET-osteo-C0-survival.tsv'

surv <- read.table(pth,
                  sep='\t',
                  header=T)

surv <- dplyr::as_tibble(surv)
```

```r
surv
```

```r
fit <- survfit(Surv(OS, vital) ~ cluster, data=surv)
g <- ggsurvplot(fit, conf.int=F, pval=T, risk.table=TRUE)
ggsave( "../img/Tree-TARGET-osteo-C0-survival-OS-survival.png", 
       print(g), 
       dpi = 1000 )
g
```
