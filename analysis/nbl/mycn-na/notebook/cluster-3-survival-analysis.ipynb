```r
library(dplyr)
library(survival)
library(survminer)
```

```r
pth <- '../data/survival/mycn-na-tme-signal-subtract-cluster3-survival.tsv'

surv <- read.table(pth,
                  sep='\t',
                  header=T)

surv <- dplyr::as_tibble(surv)
```

```r
fit <- survfit(Surv(OS, vital) ~ cell.cycle, data=surv)
g <- ggsurvplot(fit, conf.int=F, pval=T, risk.table=F,
                xlab="Time (Days)",
                font.main = c(25, "plain", "black"),
                font.x=c(25, "plain", "black"),
                font.y=c(25, "plain", "black"),
                font.legend=c(25, "plain", "black"),
                font.tickslab=c(25, "plain", "black"), 
                pval.size=8)

ggsave( "../img/cluster-3-sub-cluster-immune-OS-survival.png", 
       print(g), 
       dpi = 1000 )
g
```

```r
fit <- survfit(Surv(EFS, vital) ~ cell.cycle, data=surv)
g <- ggsurvplot(fit, conf.int=F, pval=T, risk.table=TRUE)
ggsave("../img/cluster-3-sub-cluster-immune-EFS-survival.png", 
       print(g), 
       dpi = 1000)
g
```
