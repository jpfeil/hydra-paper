```python
import sys
import pandas as pd

sys.path.append('/opt/hydra/')

import library.analysis as hydra
```

```python
exp_path = '../data/treehouse-synovial-log2TPM1-V9.tsv'
```

```python
exp = pd.read_csv(exp_path, sep='\t', index_col=0)
```

```python
from scipy.stats import ttest_ind

constant = 0.05
fgseas = {}
for sample in exp.columns:
    zscore = (exp[sample] - exp.drop(sample, axis=1).mean(axis=1)) / (exp.drop(sample, axis=1).std(axis=1) + constant)
    zscore = zscore.sort_values(ascending=False)
    fgsea = hydra.n1(zscore)
    print(fgsea)
    break
```
